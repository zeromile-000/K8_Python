```python
# 딕셔너리 (key: 리스트)
data = {
    'fruits': ['apple', 'banana', 'orange', 'grape', 'pear'],
    'vegetables': ['carrot', 'broccoli', 'spinach'],
    'drinks': ['water', 'juice']
}

# 카테고리별 갯수를 저장할 딕셔너리를 만듬
category_count = {}
category_totalcount = 0

# 항목 갯수 집계
for category, items in data.items():
    # 카테고리별 갯수를 집계
    category_count[category] = len(items)

    # 카테고리들의 총 갯수를 집계
    category_totalcount = category_totalcount + len(items)

# 결과 출력
print("카테고리별 항목 갯수:", category_count)
print("카테고리 총 항목수",category_totalcount)
```
