```python
import numpy as np
import matplotlib.pyplot as plt
############### 
### plt.title('군집 산포도') 출력을 위한 한글 폰트 설정
# 한글 폰트 경로 설정 (예: 맑은 고딕)
#font_path = "C:/Windows/Fonts/malgun.ttf"  # Windows의 경우
# 폰트 설정
#font_prop = fm.FontProperties(fname=font_path, size=12)
#plt.rc('font', family=font_prop.get_name())
#############

# 임의의 군집(클러스터) 데이터를 생성하는 함수 => 이 함수를 이해하는 것이 중요
def generate_cluster(center, num_points, spread):
    # 정규 분포를 따르는 점들을 생성
    points = np.random.randn(num_points, 2) * spread + center # 난수를 num_points 갯수 만큼, 2차원으로 생성
    return points
# np.random.randn(d0, d1, ..., dn)은 주어진 차원(d0, d1, ..., dn)의 배열을 생성
# num_points: 배열의 첫 번째 차원의 크기, 즉 생성할 데이터 포인트의 개수
# 2: 배열의 두 번째 차원의 크기, 즉 각 데이터 포인트의 차원

# 각 군집의 중심, 점 개수, 확산 정도 설정  분포도 단위 (0.1 모여있다 , 0.9 퍼져있다)
cluster_1 = generate_cluster(center=[2, 3], num_points=100, spread=0.9)
cluster_2 = generate_cluster(center=[7, 8], num_points=200, spread=0.8)
cluster_3 = generate_cluster(center=[5, 2], num_points=300, spread=0.3)

print(cluster_1[:5])

# 각 군집을 하나의 배열로 합침
all_points = np.vstack((cluster_1, cluster_2, cluster_3))
# NumPy의 vstack 함수를 사용하여 여러 배열을 수직으로 쌓아 새로운 배열을 생성하는 함수
# 군집을 시각화
plt.figure(figsize=(8, 6))

# 첫 번째 군집을 파란색으로
plt.scatter(cluster_1[:, 0], cluster_1[:, 1], color='blue', label='Cluster 1', alpha=0.6)

# 두 번째 군집을 빨간색으로
plt.scatter(cluster_2[:, 0], cluster_2[:, 1], color='red', label='Cluster 2', alpha=0.6)

# 세 번째 군집을 녹색으로
plt.scatter(cluster_3[:, 0], cluster_3[:, 1], color='green', label='Cluster 3', alpha=0.6)

# 그래프 설정
plt.title('군집 산포도')
plt.xlabel('X-축')
plt.ylabel('Y-축')
plt.legend()
plt.grid(True)

# 그래프 출력
plt.show()
```

```python
from random import choice  # 랜덤한 선택을 위해 choice 함수 임포트

class RandomWalk:
    """무작위 경로를 생성하는 클래스입니다."""

    def __init__(self, num_points=5000):
        """경로의 속성을 초기화합니다."""
        self.num_points = num_points  # 경로의 총 점 개수 지정 (기본값: 5000)

        # 모든 경로는 (0, 0)에서 시작합니다.
        self.x_values = [0]  # x 좌표 리스트 (시작점 0)
        self.y_values = [0]  # y 좌표 리스트 (시작점 0)

    def fill_walk(self):
        """경로의 모든 점을 계산합니다."""
        # 지정된 길이에 도달할 때까지 계속 이동합니다.
        while len(self.x_values) < self.num_points:
            ###################
            ### 동서남북 대각선 8가지 방향으로 이동하는 코드 작성
            # 이동 방향을 정의
            directions = [
                (1, 0),    # 동쪽: x좌표 증가
                (-1, 0),   # 서쪽: x좌표 감소
                (0, 1),    # 북쪽: y좌표 증가
                (0, -1),   # 남쪽: y좌표 감소
                (1, 1),    # 북동쪽: x와 y 모두 증가
                (-1, 1),   # 북서쪽: x 감소, y 증가
                (1, -1),   # 남동쪽: x 증가, y 감소
                (-1, -1)   # 남서쪽: x와 y 모두 감소
            ]
            # 8가지 방향을 random하게 선택
            d = choice([0,7])
            x_direction = choice([1, -1])
            x_step = x_direction * directions[d][1]
            
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_distance *  directions[d][1]
            
            ### 코딩할 부분
            ###############################

            # 이동할 방향과 이동 거리를 결정합니다.
            # x 방향을 선택합니다 (1 또는 -1).
            # x_direction = choice([1, -1])  # 동쪽(1) 또는 서쪽(-1) 방향 선택
            # x_distance = choice([0, 1, 2, 3, 4])  # 0에서 4까지의 이동 거리 선택
            # x_step = x_direction * x_distance  # 최종 x 이동 거리 계산

            # # y 방향을 선택합니다 (1 또는 -1).
            # y_direction = choice([1, -1])  # 북쪽(1) 또는 남쪽(-1) 방향 선택
            # y_distance = choice([0, 1, 2, 3, 4])  # 0에서 4까지의 이동 거리 선택
            # y_step = y_direction * y_distance  # 최종 y 이동 거리 계산

            # 이동하지 않는 경우를 거부합니다.
            if x_step == 0 and y_step == 0:  # x와 y 모두 이동 거리가 0이라면
                continue  # 다음 반복으로 넘어갑니다.

            # 새로운 위치를 계산합니다.
            x = self.x_values[-1] + x_step  # 마지막 x 값에 x 이동 거리 추가
            y = self.y_values[-1] + y_step  # 마지막 y 값에 y 이동 거리 추가

            # 새 위치를 리스트에 추가합니다.
            self.x_values.append(x)  # 새로운 x 좌표 추가
            self.y_values.append(y)  # 새로운 y 좌표 추가
```

```python
import matplotlib.pyplot as plt
from random import choice

class RandomWalk:
    """무작위 경로를 생성하는 클래스입니다."""

    def __init__(self, num_points=5000):
        """경로의 속성을 초기화합니다."""
        self.num_points = num_points  # 경로의 총 점 개수 지정 (기본값: 5000)

        # 모든 경로는 (0, 0)에서 시작합니다.
        self.x_values = [0]  # x 좌표 리스트 (시작점 0)
        self.y_values = [0]  # y 좌표 리스트 (시작점 0)

    def fill_walk(self):
        """경로의 모든 점을 계산합니다."""
        while len(self.x_values) < self.num_points:
            # 이동 방향을 정의
            directions = [
                (1, 0),    # 동쪽
                (-1, 0),   # 서쪽
                (0, 1),    # 북쪽
                (0, -1),   # 남쪽
                (1, 1),    # 북동쪽
                (-1, 1),   # 북서쪽
                (1, -1),   # 남동쪽
                (-1, -1)   # 남서쪽
            ]
            direction = choice(directions)  # 방향 선택
            distance = choice([0, 1, 2, 3, 4])  # 이동 거리 선택

            x_step = direction[0] * distance  # x 이동 거리 계산
            y_step = direction[1] * distance  # y 이동 거리 계산

            if x_step == 0 and y_step == 0:  # 이동하지 않는 경우
                continue

            # 새로운 위치 계산
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step

            self.x_values.append(x)  # 새로운 x 좌표 추가
            self.y_values.append(y)  # 새로운 y 좌표 추가
            
            


# Keep making new walks, as long as the program is active.
while True:
    # Make a random walk.
    rw = RandomWalk(100)  # 10,000개의 점을 가지는 무작위 경로 생성
    rw.fill_walk()  # 경로 점 계산

    # Plot the points in the walk.
    plt.style.use('classic')  # 그래프 스타일 설정
    fig, ax = plt.subplots()  # 서브플롯 생성
    point_numbers = range(rw.num_points)  # 점의 번호 생성
    ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues,
               edgecolors='none', s=10)  # 경로 점을 시각화
    ax.set_aspect('equal')  # x와 y 비율을 동일하게 설정

    # Emphasize the first and last points.
    ax.scatter(0, 0, c='green', edgecolors='none', s=100)  # 시작점 강조
    ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none', s=100)  # 끝점 강조

    # Remove the axes.
    ax.get_xaxis().set_visible(True)  # x축 표시
    ax.get_yaxis().set_visible(True)  # y축 표시
    
    plt.show()  # 그래프 출력

    keep_running = input("Make another walk? (y/n): ")  # 다른 경로 생성 여부 확인
    if keep_running == 'n':  # 'n' 입력 시 반복 종료
        break 
```

```python
### 벌레 이동 코딩 문제 구현 >> 
### 체스에 0이 없어질 때까지 이동 >> 바둑판 19 x19 으로 수정하는 코딩 
import numpy as np
from random import choice
import random
import matplotlib.pyplot as plt

# 10x10 배열 생성 (초기 값은 모두 0) # 바둑판 크기 세팅
grid = np.zeros((8, 8), dtype=int)
print(grid[:5])

# 벌레의 초기 위치 (임의로 설정)
x, y = random.randint(0, 7), random.randint(0, 7)

# 벌레의 이동 방향 (동서남북 및 대각선 8방향)
directions = [
    (-1, 0), (1, 0),  # 남, 북
    (0, -1), (0, 1),  # 서, 동
    (-1, -1), (-1, 1),  # 북서, 북동a
    (1, -1), (1, 1)  # 남서, 남동
]
# 반복 실행
steps = 0
while np.any(grid == 0): # grid의 값이 0이면 반복
    # 현재 위치에 1 추가
    d = choice(range(0,8)) # d변수에 0부터 7사이의 임의의 수를 선택하여 대입
    dx, dy = directions[d] # dx, dy에 directions의 [d]번째 인덱스의 값을 대입
    
    nx,ny = x+dx,y+dy # nx, ny에 x, y의 좌표 + 이동 방향을 대입(새로운 좌표)
    if nx <0 or nx >7 or ny<0 or ny > 7 : # 만약 nx와 ny가 0보다 작고 7보다 크면 
        continue # 건너뛰기

    else :    
        x,y = nx, ny # nx, ny의 값을 x와 y에 대입(다시 반복하기 위해?) #Fixme plase ㅜ_ㅜ
        steps +=1
        grid[x,y] +=1 # grid의 현재위치에 1 추가(방문횟수 +1)

    

    
#### 구현할 코드

# 결과 출력
print(f"총 {steps} 번 이동 후 배열이 채워졌습니다.")

# 배열의 결과를 시각화 (그래프 출력)
plt.imshow(grid, cmap='viridis', interpolation='none')
plt.colorbar(label="방문횟수")
plt.title("벌레의 이동 경로: 바둑판  10x10 Grid")
plt.show()
```

```python
import random
# Die 클래스를 정의 (6면체 주사위) 주사위 1개, 다이스는 주사위 여러 개
class Die:
    def __init__(self):
        self.num_sides = 6  # 주사위 면의 수

    # 주사위를 던지는 함수 (1-6 사이 임의 값을 반환)
    def roll(self):
        return random.randint(1, self.num_sides)
```
